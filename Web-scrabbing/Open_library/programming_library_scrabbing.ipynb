```python
import csv
```

```python
request= requests.get('https://openlibrary.org/search?q=programming+concepts&mode=everything')
```

```python
soup=bs(request.content,'html.parser')
soup
```

```python
containers =soup.find_all('li',class_='searchResultItem')
```

```python
for container in containers :

  title =container.find_all('h3',itemprop='name')
  title =title[0].text.strip()

  Author=container.find_all('span',itemprop='author')
  Author=Author[0].text.strip()


  import re
  year=container.find_all('span',class_='publishedYear')
  text=year[0].text.strip()
  Year=re.findall(r'\d{4}', text)[0]



  print (title +'\n' +Author +'\n' + Year +'\n' )
```

```python
with open('Open_library','w',encoding='utf-8') as f:
  obj=csv.writer(f)
  obj.writerow(['title', 'Author','Year' ] )
  for container in containers :

    title =container.find_all('h3',itemprop='name')
    title =title[0].text.strip()

    Author=container.find_all('span',itemprop='author')
    Author=Author[0].text.strip()


    import re
    year=container.find_all('span',class_='publishedYear')
    text=year[0].text.strip()
    Year=re.findall(r'\d{4}', text)[0]




  obj.writerow( [title, Author,Year ] )

  print("done")
```
